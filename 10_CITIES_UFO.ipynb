```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("ufo_sightings.csv", low_memory=False)

city_counts = df["city"].value_counts()

top10_cities = city_counts.head(10)


plt.figure(figsize=(10,6))
top10_cities.plot(kind="bar")

plt.title("Top 10 Cities with Most UFO Sightings")
plt.xlabel("City")
plt.ylabel("Number of Sightings")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 10 cities on a world map using their latitude and longitude coordinates

import plotly.express as px

top10 = df[df["city"].isin(top10_cities.index)]

fig = px.scatter_geo(
    top10,
    lat="latitude",
    lon="longitude",
    hover_name="city",
    title="Top UFO Sightings Cities"
)

fig.show()
```
