```python
import pandas as pd
import matplotlib.pyplot as plt

fp = pd.read_csv("fraud_patterns.csv")

fp_sorted = fp.sort_values("fraud_share_pct", ascending=True)

plt.barh(fp_sorted["fraud_pattern"], fp_sorted["fraud_share_pct"], color="crimson")

plt.title("Share of each type of fraud (%)")
plt.xlabel("Percentage (%)")
plt.tight_layout()
plt.show()
```
