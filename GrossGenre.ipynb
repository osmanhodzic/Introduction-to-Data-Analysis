```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("imdb_top_1000.csv", low_memory=False)

df["Gross_clean"] = df["Gross"].str.replace(",", "")
df["Gross_clean"] = pd.to_numeric(df["Gross_clean"], errors="coerce")

zanrovi = ["Action", "Drama", "Comedy", "Crime", "Adventure"]

avg_gross = {}

for z in zanrovi:
    avg = df[df["Genre"].str.contains(z, na=False)]["Gross_clean"].mean()
    avg_gross[z] = avg

avg_gross = pd.Series(avg_gross)

avg_gross.sort_values(ascending=False).plot(kind="bar", color="orange")

plt.xlabel("Genre")
plt.ylabel("Average Gross")
plt.title("Which genre earns the most?")
plt.xticks(rotation=45)

plt.show()
```
