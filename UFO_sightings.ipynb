```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("ufo_sightings.csv",low_memory=False)

df["datetime"] = pd.to_datetime(df["datetime"], errors="coerce")

df = df.dropna(subset=["datetime"])

df["year"] = df["datetime"].dt.year
df["month"] = df["datetime"].dt.month
df["hour"] = df["datetime"].dt.hour

year_counts = df["year"].value_counts().sort_index()

shape_counts = df["shape"].value_counts()

shape_counts.head(10).plot(kind="bar")
plt.title("Most Common UFO Shapes")
plt.show()

shape_country = df.groupby(["country", "shape"]).size().unstack().fillna(0)
top_countries = df["country"].value_counts().head(5).index
shape_country = shape_country.loc[top_countries]

import matplotlib.pyplot as plt

shape_country.plot(kind="bar", stacked=True)

plt.title("UFO Shapes by Country")
plt.xlabel("Country")
plt.ylabel("Number of Sightings")

year_country = df.groupby(["year","country"]).size().unstack().fillna(0)
year_country.plot()

plt.title("UFO Sightings per Year by Country")
plt.xlabel("Year")
plt.ylabel("Number of Sightings")

plt.show()

plt.show()
```
