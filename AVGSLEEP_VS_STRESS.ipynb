```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv", low_memory=False)

avg_sleep_by_stress = df.groupby('Stress Level')['Quality of Sleep'].mean()

plt.figure(figsize=(10,6))
avg_sleep_by_stress.plot(kind='bar', color='yellow')
plt.title('Stress Level + Quality of Sleep')
plt.xlabel('Stress Level')
plt.ylabel('Quality of Sleep')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
