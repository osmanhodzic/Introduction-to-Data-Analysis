```python
# Females older than 30 and stress level > 7

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv",low_memory=False)

# print(df.head())
# print(df.info())

filtered_df = df[
    (df["Gender"] == "Female") &
    (df["Age"] > 30) &
    (df["Stress Level"] > 7)
]

# print(filtered_df)

# print(filtered_df[["Gender", "Age", "Sleep Duration", "Stress Level"]])

avg_sleep_filtered = filtered_df.groupby('Age')['Sleep Duration'].mean()

plt.figure(figsize=(10,8))
avg_sleep_filtered.plot(kind='bar', color='salmon')
plt.title('Average sleep duration (women over 30, stress level > 7)')
plt.xlabel('Years')
plt.ylabel('Sleep Duration (hours)')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```
