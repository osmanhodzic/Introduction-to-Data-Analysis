```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv", low_memory=False)

engineers = df[df["Occupation"] == "Engineer"]
others = df[df["Occupation"] != "Engineer"]

avg_engineers_sleep = engineers["Sleep Duration"].mean()
avg_others_sleep = others["Sleep Duration"].mean()

high_steps = df[df["Daily Steps"] > df["Daily Steps"].mean()]
low_steps = df[df["Daily Steps"] <= df["Daily Steps"].mean()]

avg_high_steps_sleep = high_steps["Sleep Duration"].mean()
avg_low_steps_sleep = low_steps["Sleep Duration"].mean()

plt.figure(figsize=(6,4))
plt.bar(["Engineers", "Others"], [avg_engineers_sleep, avg_others_sleep], color=["orange", "blue"])
plt.title("San: Inzenjeri vs Ostali")
plt.ylabel("Sleep Duration")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

plt.figure(figsize=(6,4))
plt.bar(["High Steps", "Low Steps"], [avg_high_steps_sleep, avg_low_steps_sleep], color=["red", "green"])
plt.title("San u odnosu na broj koraka")
plt.ylabel("Sleep Duration")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()



bmi_sleep = df.groupby("BMI Category")["Quality of Sleep"].mean()


plt.figure(figsize=(7,4))
bmi_sleep.plot(kind='bar', color="teal")

plt.title("Kvalitet sna po BMI kategoriji")
plt.xlabel("BMI kategorija")
plt.ylabel("Prosjecna kvaliteta sna")

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
