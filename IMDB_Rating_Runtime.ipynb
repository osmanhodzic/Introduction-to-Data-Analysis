```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("imdb_top_1000.csv", low_memory=False)

df["Runtime"] = df["Runtime"].str.replace(" min", "").astype(int)

correlation = df["Runtime"].corr(df["IMDB_Rating"])

plt.scatter(df["Runtime"], df["IMDB_Rating"])
plt.xlabel("Runtime")
plt.ylabel("IMDB Rating")
plt.title("Are longer films rated better?")
plt.show()
```
