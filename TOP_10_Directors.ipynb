```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("imdb_top_1000.csv",low_memory=False)

top_directors = df["Director"].value_counts().head(10)

top_directors.plot(kind="barh", color="steelblue")
plt.title("Top 10 directors in the IMDB Top 1000")
plt.xlabel("Number of Movies")
plt.ylabel("Director")
plt.tight_layout()
plt.show()
```
